```python
file  = open("dfwords.txt" ,"r+")#The text in the file is copied from VocabularyWords.pdf
data = file.readlines()
```

```python
alphabets = 'abcdefghijklmnopqrstuvwxyz#'
current_alphabet = 'a'
```

```python
#removing the exersices at the end
idx = data.index("Quiz 37 (Sentence Completions)\n")
data = data[:idx]
```

Just trying to get rid of the irrelevant in between. Things like Quizes etc 

```python
ids = []

for i in range(len(data)):
    line = data[i]
    if len(line) > 0:
        if line[0].isupper() and len(line) == 2: #if a single Capital letter e.g 'B'
            
            if line[0].lower() == alphabets[alphabets.index(current_alphabet)+1]: #if the letter is the next letter in the order
                current_alphabet = line[0].lower() #change the current alphabet to the new one
                ids.append(i)
            else:
                ids.append(i) #otherwise save the idex to delete later
            
        elif line[0].isupper() and len(line) != 2  and line[0].lower() != current_alphabet: #if a random Captial word
            ids.append(i) #save to delete later    
            
        elif line[0] not in alphabets: 
            ids.append(i)
            
        elif line[0].lower() != current_alphabet: #if the word is out of order
            data[i-1] = data[i-1][:-1] + ' '+ line #it must be a continuation of the prev defination
            ids.append(i)
        
        
```

```python
for idss in sorted(ids, reverse=True): #deleting in reverse so that indices don't get changed
    del data[idss]
```
